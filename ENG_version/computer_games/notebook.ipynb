# Study of the patterns that determine the success of the game

**Brief description:** There is an online store “Strimchik” that sells computer games all over the world. Historical data on game sales, user and expert ratings, genres and platforms (for example, Xbox or PlayStation) are available from open sources. Data available up to 2016.

**Project goal:** It is necessary to identify the patterns that determine the success of the game. This will allow you to place a bet on a potentially popular product and plan advertising campaigns.

**Description of data:**

- ```name``` — the name of the game
- ```platform``` - platform
- ```year_of_release``` - year of release
- ```genre``` - game genre
- ```na_sales``` - sales in North America (millions of copies sold)
- ```eu_sales``` - sales in Europe (millions of copies sold)
- ```jp_sales``` - sales in Japan (millions of copies sold)
- ```other_sales``` - sales in other countries (millions of copies sold)
- ```critic_score``` — critics’ score (maximum 100)
- ```user_score``` — user score (maximum 10)
- ```rating``` - rating from the ESRB (Entertainment Software Rating Board). This association determines the rating of computer games and assigns them an appropriate age category.

**Decoding of ratings from the ESRB organization:**
- "EC" ("Early childhood") - "For young children": The game is suitable for children aged 3 years and above and does not contain material that parents might consider inappropriate. Products receiving this rating are primarily designed for children and are usually educational games.
- “E” (“Everyone”) - “For everyone”: The content is quite suitable for all ages (previously from 6 years). Adults can also enjoy these games. Games with this rating may contain minimal violence, mostly of a cartoon nature.
- “E10+” (“Everyone 10 and older”) - “For everyone 10 years and older”: Projects with this rating may contain more cartoon or mild violence, or somewhat explicit scenes, or a minimal amount of blood.
- “T” (“Teen”) - “Teenagers”: The game is suitable for persons over 13 years old. Projects in this category may contain violence, obscene scenes, crude humor, moderately explicit sexual content, blood, or occasional use of profanity.
- “M” (“Mature”) - “For Adults”: Game materials are not suitable for persons under 17 years of age. Projects with this rating may contain severe violence, large amounts of blood and dismemberment, obscene sexual scenes or strong profanity.
- “AO” (“Adults Only 18+”) - “Adults Only”: The contents of the game are only for adults over 18 years of age. Products in this category may contain prolonged scenes of intense violence and/or very explicit sexual content.
- “RP” (“Rating Pending”): The product has been submitted to the ESRB and is awaiting rating.
- "RP 17+" ("Rating Pending") - "Rating Pending - Approximately 17+": Product has been submitted to the ESRB for evaluation, but the approximate content is close to an "M" rating.

````python
def create_hist_for_columns(data):
    for column in data:
        try:
            if data[column].dtype not in ['float64', 'int64', 'int32', 'float32']:
                continue
            data[column].plot(kind='hist', title=f'Histogram for column {column}',
                              ylabel='Frequency'
                             )
            plt.show()
        except:
            pass
````

````python
def create_boxplot_for_columns(data):
    for column in data:
        try:
            if data[column].dtype not in ['float64', 'int64', 'int32', 'float32']:
                continue
            data[column].plot(kind='box', title=f'Boxplot for column {column}')
            plt.show()
        except:
            pass
````

### Sales Analysis from Estimates

````python
def create_graf_corr_user(platform):
    (
        actual_games.query('platform == @platform')
        .plot(kind='scatter', x='user_score', y='all_sales', alpha=0.5, figsize=(10, 7),
              title=f'Scatter plot of user ratings and game sales for the platform {platform}'
             )
    )
    return (
        actual_games.query('platform == @platform')['user_score']
        .corr( actual_games[actual_games['platform'] == platform]['all_sales'])
    )
````

````python
def create_graf_corr_critic(platform):
    (
        actual_games.query('platform == @platform')
        .plot(kind='scatter', x='critic_score', y='all_sales', alpha=0.5, figsize=(10, 7),
              title=f'Scatter plot of critical ratings versus platform game sales {platform}'
             )
    )
    return (
        actual_games.query('platform == @platform')['critic_score']
        .corr( actual_games[actual_games['platform'] == platform]['all_sales'])
    )
````

## Statistical data analysis

### Hypothesis testing 1. Are the user ratings of the Xbox One and PC platforms the same?

Let us consider the distributions for these quantities.

````python
(
    actual_games
    .query('platform == "PC"')['user_score']
    .plot(kind='hist', alpha=0.5, label='PC', legend=True)
)

(
    actual_games
    .query('platform == "XOne"')['user_score']
    .plot(kind='hist', alpha=0.5, label='XOne', legend=True)
)

plt.show()
````

The distributions are similar to normal distributions, but are skewed to the right.

The spread of the data is different.

**Null Hypothesis:** The means of user ratings for both platforms are equal.

**Alternative Hypothesis:** The average for user ratings of the platforms are not equal.

````python
alpha = 0.05

results = st.ttest_ind(
    actual_games.query('platform == "PC" and not user_score.isna()')['user_score'],
    actual_games.query('platform == "XOne" and not user_score.isna()')['user_score'],
    equal_var=False)

print('Pvalue: ', results.pvalue)

if results.pvalue < alpha:
    print('Rejecting the null hypothesis')
else:
    print('There is no reason to reject the null hypothesis')
````

We have no reason to reject the null hypothesis and are inclined to believe that the average value for user ratings of the PC and XBox One platforms are equal.

Or ordinary users give similar ratings to games for the PC and XBox One platforms.
