# Исследование объявлений о продаже квартир

**Краткое описание:** Есть данные сервиса Яндекс Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. 

**Цель проекта:** Провести исследовательский анализ данных, который поможет установить параметры, влияющие на цену объектов. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность.

**Описание данных:**
- ```airports_nearest``` — расстояние до ближайшего аэропорта в метрах (м)
- ```balcony``` — число балконов
- ```ceiling_height``` — высота потолков (м)
- ```cityCenters_nearest``` — расстояние до центра города (м)
- ```days_exposition``` — сколько дней было размещено объявление (от публикации до снятия)
- ```first_day_exposition``` — дата публикации
- ```floor``` — этаж
- ```floors_total``` — всего этажей в доме
- ```is_apartment``` — апартаменты (булев тип)
- ```kitchen_area``` — площадь кухни в квадратных метрах (м²)
- ```last_price``` — цена на момент снятия с публикации
- ```living_area``` — жилая площадь в квадратных метрах (м²)
- ```locality_name``` — название населённого пункта
- ```open_plan``` — свободная планировка (булев тип)
- ```parks_around3000``` — число парков в радиусе 3 км
- ```parks_nearest``` — расстояние до ближайшего парка (м)
- ```ponds_around3000``` — число водоёмов в радиусе 3 км
- ```ponds_nearest``` — расстояние до ближайшего водоёма (м)
- ```rooms``` — число комнат
- ```studio``` — квартира-студия (булев тип)
- ```total_area``` — общая площадь квартиры в квадратных метрах (м²)
- ```total_images``` — число фотографий квартиры в объявлении

**План работы**
1. Изучить данные
2. Предобработать данные
3. Добавить необходимые для анализа столбцы
4. Проанализировать данные
5. Оформить выводы

## Откроем файл с данными и изучим общую информацию. 

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
try:
    data = pd.read_csv('.\\datasets\\real_estate_data.csv', sep='\t', decimal='.')
except:
    data = pd.read_csv('https://code.s3.yandex.net/datasets/real_estate_data.csv', sep='\t', decimal='.')
````

````python
data.head()
````

````python
data.info()
````

````python
def create_hist_for_columns(data):
    for column in data:
        try:
            if data[column].dtype not in ['float64', 'int64', 'int32', 'float32']:
                continue
            data[column].plot(kind='hist', title=column)
            plt.show()
        except:
            pass
create_hist_for_columns(data)
````

Данные были сохранены в переменную и изучены. В данных имеются пропуски и выбивающиеся значения, некоторые стоблцы имеют некорректный для их значений тип данных. Данные необходимо предобработать.

## Предобработка данных

### Пропуски данных

Начнем предобрабатывать столбец ```ceiling_height```.

````python
data[data['ceiling_height'].isna()]
````

````python
data['ceiling_height'].describe()
````

В столбце ```ceiling_height``` имеется много пропусков и есть выбросы, данные с пропущенными значениями выглядят обычно, закономерностей не видно, поэтому принято решение заполнить пропуски медианным значением.

````python
data['ceiling_height'].fillna(data['ceiling_height'].median(), inplace=True)
````

Далее перейдем к столбцу ```living_area```.
Так как нам известны все данные столбца ```total_area```, мы найдем медианное значение того, какой процент имеет ```living_area``` от ```total_area```, и заполним все пропущенные столбцы таким образом.

````python
data['living_area'].describe()
````

````python
avg_living_to_total = (data[~data['living_area'].isna()]['living_area'] / data[~data['living_area'].isna()]['total_area']).median()
avg_living_to_total
````

````python
data['living_area'].fillna(data['total_area'] * avg_living_to_total, inplace=True)
````

По такому же принципу поступим с ```kitchen_area```

````python
avg_kitchen_to_total = (data[~data['kitchen_area'].isna()]['kitchen_area'] / data[~data['kitchen_area'].isna()]['total_area']).median()
avg_kitchen_to_total
````

````python
data['kitchen_area'].fillna(data['total_area'] * avg_kitchen_to_total, inplace=True)
````

Заполним все пропуски столбца ```balcony``` 0, так как если продавец не ввел данные о балконе, то скорее всего его нет.


````python
data['balcony'].fillna(0, inplace=True)
````

По такому же принципу заполним пропуски в столбце ```is_apartment```. Если данные не указаны, скорее всего это не апартаменты.

````python
data['is_apartment'].fillna(False, inplace=True)
````

````python
data.info()
````

### Приведение типов данных

Приведем столбец ```first_day_exposition``` к типу данных datetime.

````python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
data['first_day_exposition'].head()
````

Столбец ```balcony``` приведем к типу int, так как балконов не может быть дробное количество.
```floors_total``` и ```days_exposition``` имеют пропущенные значения, поэтому первести к типу данных int мы не можем.

````python
data['balcony'] = data['balcony'].astype('int64', errors='ignore')
````

````python
data.info()
````

### Дубликаты

Неявные дубликаты возможны только в столбце ```locality_name```. Проверим его.

````python
data['locality_name'].sort_values().unique()
````

````python
data['locality_name'].nunique()
````

````python
data['locality_name'] = (
    data['locality_name']
    .str.replace('ё', 'е')
    .str.replace('городской поселок', 'поселок городского типа')
    .str.replace('поселок станции', 'посёлок при железнодорожной станции')
    .str.replace('садоводческое некоммерческое товарищество', 'садовое товарищество')
)
````

````python
data['locality_name'].sort_values().unique()
````

````python
data['locality_name'].nunique()
````

### Аномальные значения

Если просмотреть гистограммы, посроенные для обзора данных, можно увидеть аномалии в столбце ```ceiling_height```, так же мы строили числовое описание этому столбцу, где было видно, что в столбце есть объекты, у которых высота потолка = **100 метров**. Невероятная высота при учете, что высокими потолками считаются потолки высотой около 3 метров. Посмотрим на данные, где потолки выше или равны 10 метрам.

````python
data[data['ceiling_height'] >= 10]['ceiling_height']
````

Можно заметить, что высота потолков 25-32 метров возможно была неправильно внесена в базу данных, и это должны быть значения 2.5-3.2 метров, что выглядит куда более реалистично. Исправим эти значения, а аномально высокие потолки 100 или 10 метров мы удалим из датасета.

````python
data['ceiling_height'].where(data['ceiling_height'] < 10, data['ceiling_height'] / 10, inplace=True)
````

````python
data['ceiling_height'].hist()
````

После обработки у нас также остались выбивающиеся значения, где высота потолка составляет 4.5 и более метров или менее 2 метров.

Удалим такие значения из датасета, так как они являются большим выбросом.

````python
data = data.query('ceiling_height < 4.5 and ceiling_height > 2')
````

Посмотрим на гистограммы для обновленного датасета.

````python
create_hist_for_columns(data)
````

Так же по гистограммам можно увидеть большой выброс в столбце ```last_price```. Изучим его.

Удалим из датасета все квартиры, стоимость которых составляет более 20 млн., так как большинстов квартир идут в одном ценовом сегменте, но очень дорогие квартиры, которые включены в другой ценовой сегмент(напрмиер премиум) будут мешать нам оценивать рыночную стоимость среднестатистических квартир.

### Средняя цена одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений

````python
top_10 = data['locality_name'].value_counts()[:10]
top_10
````

Получем топ-10 населенных пунктов с самым большим количеством объявлений. <a id='top10'></a>

````python
(
    data
    .query('locality_name in @top_10.index')
    .pivot_table(index='locality_name', values='price_square_meter', aggfunc=['count', 'mean'])
    .sort_values(by=('mean', 'price_square_meter'), ascending=False)
)
````

Мы построили сводную таблицу из которой видно количество объявлений и среднюю цену одного квадратного метра для выбранных населенных пунктов.

Наиболее дорогие квадратные метры находятся в городе Санкт-Петербург и Пушкин, цена около 100 - 110 тысяч за м2, а наиболее дешевые в городе Всеволожск и Выборг, цена около 58 - 68 тысяч за м2.
