# Исследование определяющих успешность игры закономерностей

**Краткое описание:** Есть интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Имеются данные до 2016 года.

**Цель проекта:** Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Описание данных:**

- ```name``` — название игры
- ```platform``` — платформа
- ```year_of_release``` — год выпуска
- ```genre``` — жанр игры
- ```na_sales``` — продажи в Северной Америке (миллионы проданных копий)
- ```eu_sales``` — продажи в Европе (миллионы проданных копий)
- ```jp_sales``` — продажи в Японии (миллионы проданных копий)
- ```other_sales``` — продажи в других странах (миллионы проданных копий)
- ```critic_score``` — оценка критиков (максимум 100)
- ```user_score``` — оценка пользователей (максимум 10)
- ```rating``` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

**Расшифровка рейтингов от организации ESRB:**
- «EC» («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры.
- «E» («Everyone») — «Для всех»: Содержание вполне подходит для всех возрастов (ранее с 6 лет). Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера.
- «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»: Проекты с данным рейтингом могут содержать больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови.
- «T» («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.
- «M» («Mature») — «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет. Проекты с данным рейтингом могут содержать жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику.
- «AO» («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое.
- «RP» («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга.
- «RP 17+» («Rating Pending») — «Рейтинг ожидается - примерно 17+»: Продукт отправлен ESRB на оценку, но примерное содержание близко к рейтингу «M».

````python
def create_hist_for_columns(data):
    for column in data:
        try:
            if data[column].dtype not in ['float64', 'int64', 'int32', 'float32']:
                continue
            data[column].plot(kind='hist', title=f'Гистограмма для столбца {column}',
                              ylabel='Частота'
                             )
            plt.show()
        except:
            pass
````

````python
def create_boxplot_for_columns(data):
    for column in data:
        try:
            if data[column].dtype not in ['float64', 'int64', 'int32', 'float32']:
                continue
            data[column].plot(kind='box', title=f'Boxplot для столбца {column}')
            plt.show()
        except:
            pass
````

## Шаг 3. Исследовательский анализ

### Поиск актуальных данных

Посмотрим сколько игр выпустилось в разные годы.

````python
(
    games['year_of_release']
    .value_counts()
    .sort_index()
    .plot(kind='bar', figsize=(10, 6), grid=True,
          title='Сколько игр выпущено в каждый год',
          xlabel='Год'
         )
)
````

Больше всего игр было выпущено в 2008 и 2009 годах, менее 200 игр в год выпускалось в 1980 - 1995 года.

Посмотрим сколько было выпущено игр по годам для каждой платформы.

````python
for platform in games['platform'].unique():
    (games
     .query('platform == @platform')['year_of_release']
     .value_counts()
     .sort_index().plot(kind='bar',
                          title=f'Сколько игр выпущено в каждый год для платформы {platform}',
                          xlabel='Год'
                         )
    )
    plt.show()
````

Посмотрим на продажи игр для каждой платформы.

````python
sales_for_platforms = games.groupby('platform')['all_sales'].sum()
sales_for_platforms.sort_values().plot(kind='bar', figsize=(8, 5),
                                       title='Продажи для платформ (млн.)',
                                       xlabel='Платформы', ylabel='Кол-во проданных копий (млн.)'
                                      )
````

Отберем платформы, чьи продажи составляют более 400 млн. и узнаем за какой характерный срок появляются новые и исчезают старые платформы.

Для начала посмотрим на распределение продаж игр для каждой такой платформы.

````python
for platform in sales_for_platforms[sales_for_platforms >= 400].index:
    (games
     .query('platform == @platform')
     .groupby('year_of_release')['all_sales'].sum()
     .plot(kind='bar',
           title=f'Распределение продаж для платформы {platform}',
           xlabel='Год', ylabel='Продажи'
          )
    )
    plt.show()
````

Проанализировав полученные графики можно судить, что своего пика платформы достигают за 4 года в среднем, после данного периода у платформа начинается спад в продажах.

Возьмем за актуальные данные все игры, чей релиз не был не более 4 лет назад. 2017 год - 4 года, получается их год выпуска должен быть не меньше 2013 года.

````python
actual_games = games[games['year_of_release'] >= 2013]
actual_games.describe()
````

### Поиск потенциально прибильных платформ

Рассмотрим графики, на которых будут изображены количество выпущенных игр(синий столбец) и продаж копий (млн.) (оранжевый столбец) по годам для каждой платформы.

````python
for platform in actual_games['platform'].unique():
    (
        actual_games
        .query('platform == @platform')
        .groupby('year_of_release')['all_sales'].agg(['count', 'sum'])
        .plot(kind='bar', title=f'График количества выпущенных игр и продаж за год для платформы {platform}',
              xlabel='Год', legend=True
             )
    )
    plt.legend(['Проданных игр', 'Продаж'])
````

В качестве потенциально прибильных платформ я могу выделить:
- PS4
- PC
- XOne

Выделяем по следующему ряду причин:
- для данных платформ имеется тенденция к росту по количеству выпускаемых игр в последние года;
- имеются хорошие продажи для данных платформ

Построим boxplot для каждой платформы по глобальным продажам.

````python
plt.figure(figsize=(10, 6))
sns.boxplot(data=actual_games, x='platform', y='all_sales')
plt.title('Продажи разных платформ')
plt.xlabel('Платформы')
plt.ylabel('Продажи')
````

На графике выше мы видим, что для всех платформ продажи в среднем находятся около 0. Так же у популярных платформ имеется много выбросов, которые означают аномально высокие продажи для этих платформ.

Самые большие выбросы имеют следующие платформы:
- PS3
- X360

### Анализ продаж от оценок

````python
def create_graf_corr_user(platform):
    (
        actual_games.query('platform == @platform')
        .plot(kind='scatter', x='user_score', y='all_sales', alpha=0.5, figsize=(10, 7),
              title=f'Точечный график зависимости оценок пользователей и продаж игр для платформы {platform}'
             )
    )
    return (
        actual_games.query('platform == @platform')['user_score']
        .corr( actual_games[actual_games['platform'] == platform]['all_sales'])
    )
````

````python
def create_graf_corr_critic(platform):
    (
        actual_games.query('platform == @platform')
        .plot(kind='scatter', x='critic_score', y='all_sales', alpha=0.5, figsize=(10, 7),
              title=f'Точечный график зависимости оценок критиков и продаж игр для платформы {platform}'
             )
    )
    return (
        actual_games.query('platform == @platform')['critic_score']
        .corr( actual_games[actual_games['platform'] == platform]['all_sales'])
    )
````

## Статистический анализ данных

### Проверка гипотезы 1. Пользовательские рейтинги платформ Xbox One и PC одинаковые?

Рассмотрим распределения для этих величин.

````python
(
    actual_games
    .query('platform == "PC"')['user_score']
    .plot(kind='hist', alpha=0.5, label='PC', legend=True)
)

(
    actual_games
    .query('platform == "XOne"')['user_score']
    .plot(kind='hist', alpha=0.5, label='XOne', legend=True)
)

plt.show()
````

Распределения похожи на нормальные распределения, но скошены вправо.

Разброс у данных разный.

**Нулевая гипотеза:** средние значений для пользовательских оценок обоих платформ равны.

**Альернативная гипотеза:** среднее значение для пользовательских оценок платформ не равны.

````python
alpha = 0.05

results = st.ttest_ind(
    actual_games.query('platform == "PC" and not user_score.isna()')['user_score'],
    actual_games.query('platform == "XOne" and not user_score.isna()')['user_score'],
    equal_var=False)

print('Pvalue: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергать нулевую гипотезу')
````

У нас нет основания отвергнуть нулевую гипотезу и мы склоняемся к тому, что среднее значение для пользовательских оценок платформы PC и XBox One равны.

Или обычные пользователи ставят схожие оценки играм для платформы PC и XBox One.

### Проверка гипотезы 2. Пользовательские рейтинги жанров Action и Sports разные?

Рассмотрим на распределения для этих величин.

````python
(
    actual_games
    .query('genre == "Action"')['user_score']
    .plot(kind='hist', alpha=0.5, label='Action', legend=True, bins=15)
)

(
    actual_games
    .query('genre == "Sports"')['user_score']
    .plot(kind='hist', alpha=0.5, label='Sports', legend=True, bins=10)
)

plt.show()
````

Распределения имеют нормальный вид, скошены вправо, размер и  разброс у данных разный.

**Нулевая гипотеза:** средние значений для пользовательских оценок обоих жанров равны.

**Альернативная гипотеза:** средние значения для пользовательских оценок для обоих жанров не равны.

````python
alpha = 0.05

results = st.ttest_ind(
    actual_games.query('genre == "Action" and not user_score.isna()')['user_score'],
    actual_games.query('genre == "Sports" and not user_score.isna()')['user_score'],
    equal_var=False)

print('Pvalue: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергать нулевую гипотезу')
````

По результатам статистического теста у нас есть основания отвергнуть нулевую гипотезу и склонится к тому, что средние значения для пользовательских оценок для обоих жанров разные.
